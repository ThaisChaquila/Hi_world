```python
import pandas as pd
import altair as alt
```

```python
esperanza_vida = pd.read_csv("DP_LIVE_05072021045449693.csv")
```

```python
esperanza_vida.columns = esperanza_vida.columns.str.lower()
esperanza_vida
```

```python
# limpieza de columnas
esperanza_vida.drop(columns = 'flag codes', inplace = True)
esperanza_vida
```

```python
# se busca saber los valores máximos y mínimos del value
esperanza_vida.sort_values("value", ascending = False, inplace = True)
esperanza_vida
```

### Primer gráfico:

```python
# primer gráfico: esperanza de vida de mujeres del año 2019
cond_1 = esperanza_vida['subject'] == 'WOMEN'
cond_2 = esperanza_vida['time'] == 2019

filtro_1 = esperanza_vida[cond_1 & cond_2]
```

```python
filtro_1
```

## Esperanza de vida de las mujeres en el año 2019

**Análisis:**

Según, la Organización Mundial de la Salud, las mujeres siempre han vivido mas años que los hombres. Se afirma que en promedio las mujeres sobrepasan al grupo masculino por 1,4 años. Existen diversos factores que evidencian este hecho, de los cuales uno es la actitud frente a exámenes médicos. Se demuestra que las mujeres son mas dispuestas a realizarse pruebas o seguir terapias. Además, el porcentaje de suicidios, los accidentes automovilísticos y los homicidios son 4 veces más alta que las mujeres. Sin embargo, la expectativa de vida de las mujeres disminuye si carecen de acceso a los servicios de salud y tiene bajos ingresos económicos, lo que sucede en los países de desarrollo (Noticias ONU, 2019).


Bajo este panorama y la base de datos seleccionados de la OECD, se infiere que en el 2019 Japón, un país desarrollado, tuvo un índice de 87 con respecto a la esperanza de vida de las mujeres. Sin embargo, lo mismo no sucedió con Sudáfrica, ya que se le asignó un índice de 67.7 el cuál lo lleva a representar el país con menor índice.

```python
graf1 = alt.Chart(filtro_1,width = 900, height = 500).mark_point().encode(
    alt.X("location:O", title = "Ubicación"),
    alt.Y("value:Q", title = "Valor"),
    color = alt.Color("value",scale = alt.Scale(scheme = 'turbo')),
    tooltip=["location","value"]
).interactive().properties(
    title = {
    "text" : ["Esperanza de vida de mujeres en el año 2019"],
        "subtitle" : ["Fuente: OECD (2021), Life expectancy at birth (indicator)"],
        "color" : "Brown",
        "subtitleColor" : "Black"
    })

graf1.configure_title(
    fontSize = 20,
    font = 'Courier',
    anchor = 'start',
    )
```

### Segundo gráfico:

```python
# segundo gráfico: esperanza de vida de hombres y mujeres en Japón
cond_4 = esperanza_vida['location'] == 'JPN'
cond_5 = esperanza_vida['subject'] == 'TOT'
filtro_3 = esperanza_vida[cond_4 & cond_5]
```

## Esperanza de vida al nacer en Japón

**Análisis:**

La tasa de vida en Japón ha subido hasta los 84,1 años, lo que conlleva que este país asiático se encuentre en el puesto 4 de los países más longevos del mundo. Esta cifra se debe al sistema de salud que los japoneses han implementado desde la Segunda Guerra Mundial. Actualmente, ellos van al médico un promedio de 13,4 veces al año, mas del doble que los ciudadanos del Reino Unido. A esto se añade que ponen hincapié a la prevención de factores de riesgos más importantes como enfermedades crónicas, consumo de tabaco, obesidad y la tensión arterial no controlada (El universal, 2020).

Asimismo, según la base de datos de la OECD, se puede deducir que desde el año 2016 el índice de esperanza en Japón ha aumentado constantemente en 0.12% cada año hasta el 2019.

```python
graf2 = alt.Chart(filtro_3,width = 500, height = 500).mark_bar().encode(
    alt.X("time:O", title = "Año"),
    alt.Y("value:Q", title = "Valor"),
    color = alt.Color("value",scale = alt.Scale(scheme = 'set1')),
    tooltip=["location","value"]
).interactive().properties(
    title = {
        "text" : ["Esperanza de vida en Japón"],
        "subtitle" : ["Fuente: OECD (2021), Life expectancy at birth (indicator)"],
        "color" : "Olive",
        "subtitleColor" : "black"
    })

graf2.configure_title(
    fontSize = 20,
    font = 'Courier',
    anchor = 'start',
    )
```

## Esperanza de vida de los hombres en Sudáfrica

**Análisis:**

El acceso a la salud no es prioridad para los residentes donde la esperanza de vida en la más baja. Esta cuestión no solo es discutible en términos económicos o sanitarios, sino que abarca aspectos diversos en la calidad de vida de los africanos. La tasa de mortalidad baja se debe a las enfermedades como infecciones intestinales, respiratorias y malaria. Lo que significa que estas personas no morirían si vivieran en países pobres, territorio y clima que favorecen a distintas plagas y enfermedades. Lamentablemente, vivir en un país con grandes desigualdades no solo en la región, sino en el mismo país, dificulta el avance del sistema sanitario (La vanguardia, 2017).

En este contexto social y de acuerdo a la base de datos seleccionada, la OECD indica que en el 2019 el indice de vida de los hombres en Sudafrica aumentó levemente en 1.38% respecto al 2016, año inicial de estudio.

```python
graf_3 = alt.Chart(filtro_4,width = 500, height = 500).mark_square().encode(
    alt.X("time:O", title = "Año"),
    alt.Y("value:Q", title = "Valor"),
    color = alt.Color("value",scale = alt.Scale(scheme = 'tableau10')),
    tooltip=["location","value"]
).interactive().properties(
    title = {
        "text" : ["Esperanza de vida de los hombres en África"],
        "subtitle" : ["Fuente: OECD (2021), Life expectancy at birth (indicator)"],
        "subtitleColor" : "Black"
    })

graf_3.configure_title(
    fontSize = 20,
    font = 'Courier',
    anchor = 'start',
    color = 'gray')
```
